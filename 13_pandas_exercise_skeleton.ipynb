## Q14.
<div class="alert alert-success">
[문제14]     
    
보험료('charges')와 가장 상관관계가 높은 변수는?
</div>

[Hint]
 - 공분산: df.cov()
 - 상관계수: df.corr()

```python
columns = list(df.columns)
 
num_df = df.select_dtypes(include='number')

# 상관계수 행렬
corr_matrix = num_df.corr()

# 공분산 행렬
cov_matrix = num_df.cov()

print("상관계수 행렬:")
print(corr_matrix)

print("\n공분산 행렬:")
print(cov_matrix)
```
