```python
dates=['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04',  '2024-10-05'  ]
prices=[100,102,101,105,103]
plt.figure(figsize=(10,6))
plt.title('stock Prices Over 5 Days')
plt.plot( dates,prices, marker='o',label="stock price")

plt.xlabel('Date')
plt.xlabel('Price')
plt.ylim(95, 110)  # 최소 95
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
plt.subplot(1,2, 1)
plt.plot( dates,prices) #line
plt.xticks(rotation=40)


plt.subplot(1,2, 2)
plt.bar( dates,prices) 
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart.png')
```

```python
data1 = np.random.randn(100)

plt.subplot(1,2, 1)
plt.plot(data1) #line
plt.xticks(rotation=40)


plt.subplot(1,2, 2)
plt.bar(range(len(data1)), data1, width=0.8) 
plt.xticks(rotation=40)

plt.subplot(1, 3, 2)
plt.hist(data1, bins=30, color='skyblue', edgecolor='black')  # width 대신 bins 사용
plt.xticks(rotation=40)
plt.title("3 style chart")
plt.tight_layout()
```
