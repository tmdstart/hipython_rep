```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
df=pd.read_csv('./data/titanic.csv')
df


df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# LabelEncoder 객체 생성
le = LabelEncoder()

# 'Embarked' 컬럼에 레이블 인코딩 적용
# fit_transform은 학습과 변환을 동시에 수행합니다.
df['Embarked'] = le.fit_transform(df['Embarked'])

print("\n--- 레이블 인코딩 후 데이터프레임 (Embarked 컬럼 확인) ---")
print(df[['PassengerId', 'Embarked']].head())
print("\n--- 레이블 인코딩된 값과 원본 문자열 매핑 확인 ---")
# 어떤 문자열이 어떤 숫자로 변환되었는지 확인할 수 있습니다.
for i, label in enumerate(le.classes_):
    print(f"{label}: {i}")

# 이제 X 데이터프레임 구성
X = df[['Pclass', 'Embarked', 'SibSp']]

print("\n--- 최종 X 데이터프레임 (일부) ---")
print(X.head())
print("\n--- X 데이터프레임의 정보 ---")
print(X.info())
```

```python
df['Sex'] = df['Sex'].replace({'female': 0, 'male': 1})

df
```

```python
X = df[['Pclass', 'SibSp','Sex','Parch','Embarked','Age']]
y = df['Survived']

# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(
    X, y,                 # 데이터프레임 기준
    test_size=0.2,        # 테스트셋 비율 (20%)
    random_state=42       # 재현 가능성 유지
)
```

```python
# 모델 생성
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
# 모델 학습
dt_clf.fit(X_train, y_train)
# 모델 예측
pred = dt_clf.predict(X_test)
```

```python
# 모델 평가

accuracy_score(y_test, pred)
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5)
for train_index, test_index in skf.split(iris_df, iris_df['label']):
  train_data = iris_df.iloc[train_index]
  test_data = iris_df.iloc[test_index]
  X_train = train_data[iris.feature_names]
  X_test = test_data[iris.feature_names]

  y_train = train_data['label']
  y_test =  test_data['label']

  model = DecisionTreeClassifier(random_state=156)
  model.fit(X_train, y_train)
  pred = model.predict(X_test)
  acc = accuracy_score(y_test, pred)
  print(f'acc : {acc}')
```

Alcohol (알코올): 와인에 포함된 에탄올의 비율을 나타냅니다. 발효 과정에서 포도의 당분이 효모에 의해 알코올과 이산화탄소로 변환되어 생성되며, 와인의 바디감과 맛의 강도에 영향을 미칩니다. 일반적으로 9%에서 13% 사이지만, 따뜻한 지역의 와인은 더 높을 수 있습니다.

Malic Acid (말산): 와인에 함유된 주요 산 중 하나로, 덜 익은 사과나 포도에서 주로 발견됩니다. 와인에 신선하고 상큼한 맛을 부여하며, 특히 차가운 지역의 화이트 와인에서 특징적인 산미를 제공합니다. 일부 와인 스타일에서는 말산 발효를 통해 부드럽고 조화로운 맛을 내기도 합니다.

Ash (재): 와인을 증발시키고 완전히 태운 후 남는 무기물(미네랄 및 미량 원소)의 총량을 의미합니다. 칼륨, 인, 칼슘, 마그네슘, 나트륨 등의 미네랄과 망간, 아연, 구리, 철 등의 미량 원소를 포함하며, 와인의 원산지, 토양, 포도 품종 등에 따라 함량이 달라질 수 있습니다.

Alkalinity of Ash (재의 알칼리도): 와인에 있는 유기산과 결합된 양이온(암모늄 이온 제외)의 총량을 나타냅니다. 주로 칼륨 함량과 관련이 있으며, 와인의 완충 능력과 맛에 영향을 미칠 수 있습니다.

Magnesium (마그네슘): 와인에 함유된 미네랄 중 하나입니다. 높은 농도의 마그네슘은 와인에 짠맛을 부여할 수 있으며, 전반적인 미네랄 균형에 기여합니다.

Total Phenols (총 페놀): 와인에 존재하는 모든 페놀 화합물의 총량을 의미합니다. 포도에서 유래한 이 화합물들은 와인의 색, 향, 쓴맛, 떫은맛, 그리고 노화 과정에 중요한 역할을 합니다. 또한 항산화 및 항염증 특성과 관련하여 와인의 건강상의 이점에도 기여합니다.

Flavanoids (플라보노이드): 총 페놀 중 한 종류로, 주로 포도 껍질과 씨앗에서 추출됩니다. 와인의 떫은맛과 쓴맛에 크게 기여하며, 특히 레드 와인의 색상과 산화 노화에 영향을 미칩니다. 화이트 와인의 경우 갈변 현상과 조기 산화 노화의 주요 원인이 될 수 있습니다.

Nonflavanoid Phenols (비플라보노이드 페놀): 플라보노이드 외의 페놀 화합물을 총칭합니다. 하이드록시벤조산, 하이드록시신남산, 스틸벤 등이 있으며, 비록 색이 없거나 옅은 노란색을 띠지만 레드 와인의 색상을 강화하고 안정화하는 데 기여합니다. 또한 와인의 향미에 영향을 미치고, 일부는 강력한 생물학적 활성을 나타내기도 합니다 (예: 레스베라트롤).

Proanthocyanins (프로안토시아닌): 플라반-3-올의 올리고머 및 중합체로, 포도와 와인에서 발견됩니다. 레드 와인의 색상과 입안에서의 질감(떫은맛)에 큰 영향을 미칩니다. 강력한 항산화 특성을 가지고 있어 심혈관 질환 및 암 예방과 같은 건강상의 이점과도 관련이 있습니다.

Color Intensity (색상 강도): 와인의 색깔이 얼마나 진하고 불투명한지를 나타냅니다. 일반적으로 색상 강도가 높을수록 와인의 풍미가 더 강렬하고 바디감이 있을 가능성이 높습니다.

Hue (색조): 와인의 색깔이 어떤 경향을 띠는지 나타냅니다 (예: 붉은색, 보라색, 벽돌색 등). 와인의 색조는 포도 품종, 숙성 정도, pH(산도) 등에 따라 달라지며, 와인의 나이와 특성을 파악하는 중요한 지표입니다.

OD280/OD315 of diluted wines (희석 와인의 OD280/OD315): 특정 파장(280nm 및 315nm)에서 희석된 와인의 광학 밀도(Optical Density) 비율을 나타냅니다. 이 비율은 와인에 함유된 특정 페놀 화합물의 함량과 관련이 있을 수 있으며, 와인의 특성을 분류하는 데 사용될 수 있습니다. (정확한 화학적 의미는 복잡하여 생략합니다만, 와인의 성분 구성을 반영하는 지표 중 하나입니다.)

Proline (프롤린): 와인에 존재하는 아미노산 중 하나입니다. 최근 연구에 따르면 프롤린은 레드 와인의 점성, 단맛, 풍미를 향상시키고 쓴맛과 떫은맛을 억제하는 것으로 나타났습니다. 특히 따뜻한 기후에서 재배된 포도에서 프롤린 함량이 더 높은 경향이 있습니다.
